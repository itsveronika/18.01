Присоединяю датафрейм снизу


```python
b1 = pd.concat([ptd5, ptd12], ignore_index=True)
b2 = pd.concat([ptd6, ptd10], ignore_index=True)
b3 = pd.concat([ptd2, ptd11], ignore_index=True)
b4 = pd.concat([ptd1, ptd9], ignore_index=True)
b5 = pd.concat([ptd3, ptd8], ignore_index=True)
b6 = pd.concat([ptd4, ptd7], ignore_index=True)
```

```python
b1
```

```python
df_new = b1.merge(b2, left_on='ID', right_on='ID')
df_new = df_new.merge(b3, left_on='ID', right_on='ID')
df_new = df_new.merge(b4, left_on='ID', right_on='ID')
df_new = df_new.merge(b5, left_on='ID', right_on='ID')
df_new = df_new.merge(b6, left_on='ID', right_on='ID')
df_new = df_new.sort_values(by='ID')
df_new
```

```python
df_new.shape
```

Сортирую колонку ID по возрастанию при помощи функции sort_values

```python
df1 = df_new.sort_values("ID") 
df1
```

Удаляю всевозможные дубликаты

```python
df1.drop_duplicates(keep= False )
```

Привожу текстовые значения к нижнему регистру при помощи функции str.lower
